```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [2, 2, 8, 5, 7, 6, 1, 4]
y = [10, 5, 4, 8, 5, 4, 2, 9]

plt.figure(figsize=(5, 5))
labels = [f"A{i+1}" for i in range(len(x))]
clusters = [1, 1, 2, 2, 2, 1, 3, 3]

for i, label in enumerate(labels):
    plt.text(x[i] + 0.1, y[i] + 0.1, label, fontsize=9, ha='right')

plt.scatter(x, y)
plt.grid(True)
plt.show()
```

```python
x = [2, 2, 8, 5, 7, 6, 1, 4]
y = [10, 5, 4, 8, 5, 4, 2, 9]


labels = [f"A{i+1}" for i in range(len(x))]
clusters = [1, 3, 2, 2, 2, 2, 3, 2]

colors = {1: '#FF6B6B', 2: '#4ECDC4', 3: '#556270'}
plt.figure(figsize=(5, 5))
for i in range(len(x)):
    plt.scatter(x[i], y[i], color=colors[clusters[i]], label=f"C {clusters[i]}" if f"C {clusters[i]}" not in plt.gca().get_legend_handles_labels()[1] else "")

# Etiquetas de puntos
for i, label in enumerate(labels):
    plt.text(x[i] + 0.1, y[i] + 0.1, label, fontsize=9, ha='right')


plt.legend(title="Clusters")

#plt.scatter(x, y)
plt.grid(True)
plt.show()
```
